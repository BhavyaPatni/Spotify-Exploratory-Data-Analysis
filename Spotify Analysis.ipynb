```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_tracks = pd.read_csv('tracks.csv')
```

```python
df_tracks.head(10)
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
df_tracks.describe().transpose()
```

```python
sorted_df = df_tracks.sort_values('popularity',ascending = True).head(10)
sorted_df
```

```python
most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending = False)
most_popular[:10]
```

```python
#this has set our index id to be release date instead of 0,1,2,3...

df_tracks.set_index('release_date',inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks.columns
```

```python
#this will give us artist name al index location 18 , i.e., at row number 18

df_tracks[["artists"]].iloc[18]
```

```python
#this will change milli seconds into seconds

df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms",inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
df_tracks.corr()
```

```python
temp = df_tracks[['id', 'name', 'popularity', 'duration', 'artists',
       'id_artists', 'danceability', 'energy', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']]
```

```python
temp.head()
```

```python
plt.figure(figsize=(14,8))
heatmap = sns.heatmap(temp.corr(),annot = True ,cmap="inferno")
heatmap.set_title("Correlation Heatmap Between Variable")
```

```python
#creating regression models between different variables

df_sample = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(df_sample))
```

```python
print(len(df_tracks))
```

```python
plt.figure(figsize=(14,8))
sns.regplot(data = df_sample, y= "loudness", x = "energy",color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(14,8))
sns.regplot(data = df_sample, y= "acousticness", x = "energy",color = "b").set(title = "acousticness vs Energy Correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=7,kind="hist").set(title="Songs per year")
```

```python
total_dr = df_tracks.duration
fig, ax = plt.subplots(figsize = (18,7))
fig= sns.barplot(x = years,y = total_dr, ax = ax,errwidth=False)
plt.xticks(rotation=90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig, ax = plt.subplots(figsize = (10,5))
fig= sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 60)
```
